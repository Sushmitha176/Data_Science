```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer()
data['Age']=imp.fit_transform(data[['Age']])
data
```

```python
data['Cabin']=data['Cabin'].fillna(value='unknown')
data
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode())
data
```

```python
data['Pclass'].count()
```

```python
import seaborn as sns
sns.countplot(data=data,x='Pclass')
```

```python
female_passengers=data[(data['Sex']=='female')&(data['Survived']==1)]
names=female_passengers['Name']
names
```

```python
passenger=data[(data['Pclass']==3)&(data['Age']<18)]
passenger
```

```python
old_pass=data[(data['Pclass']==1)&(data['Age']>40)]
old_pass
```

```python
combine_passenger=pd.concat([passenger,old_pass])
survived_passenger=combine_passenger[combine_passenger['Survived']==1]
survived_passenger
```

```python
male_passenger=data[(data['Sex']=='male')&(data['Fare']>100)]
male_passenger
```
